# The Overestimation Problem

- Hado V. Hasselt describes the overestimation problem in his **Double Q-Learning paper**: https://papers.nips.cc/paper/3964-double-q-learning
- This blog also gives a good explanation about the overestimation problem: https://papers.nips.cc/paper/3964-double-q-learning


## Source of overestimation
- Insufficiently flexible function approximation
- Noise or Stochasticity (in rewards and/or environment)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

## Senario 1

- This problem is introduced in Hado V. Hasselt's paper **Deep Reinforcement Learning with Double Q-learning**: https://arxiv.org/abs/1509.06461
- In this problem the true optimal action values depend only on state so that in each state all actions have the same true value
- Here we only experiment on the case where $Q^*(s,a)=2*exp(-s^2)$

```python
def true_action_value_senario_1(s):
    return 2*math.exp(-float(s)*float(s))

def plot_true_action_values():
    x = np.linspace(-6, 6, 100)
    y = np.array([true_action_value_senario_1(x_i) for x_i in x])
    plt.plot(x, y, '-r')

plot_true_action_values()
```

### We implemented the ordinary q-learning algorithm and two variants of the double q-learning
- d: the estimate is a d-degree polynomial (we use 6 and 9 here)
- n_action: the number of action at each state (we use 10 here)
- In the paper, each action has one d-degree polynomial function

### The q-learning agent

```python
class q_learning_agent_env_1():
    def __init__(self, d, n_action, learning_rate=0.5, discount_factor=1.0):
        # use n_action functions to estimate the action-state values
        self.weights = np.zeros(shape=(n_action, d))
        self.d = d
        self.n_action = n_action
        self.exp_X = dict()
        self.exp_Y = dict()
        
    def get_name(self):
        return 'q_learning_agent'
        
    def get_feature_vector(self, s):
        return np.array([s**i for i in range(0, self.d)])
    
    def get_q_value(self, s, a):
        w = self.weights[a]
        return np.dot(w, self.get_feature_vector(s))
    
    def compute_weights(self):
        X = self.exp_X
        Y = self.exp_Y
        
        for a in range(0, self.n_action):
            a_X = np.array(X[a])
            a_Y = np.array(Y[a])
            self.weights[a, :] = np.dot(np.dot(np.linalg.inv(np.dot(a_X.T, a_X)), a_X.T), a_Y)
    
    def update(self, current_s, a, next_s, reward):
        """
        In this problem, we ignore the next state
        """
        # this is a polynomial function approximation problem, we can't simply use gradient descent
        # so we store the experiences
        if a not in self.exp_X:
            self.exp_X[a] = []
            self.exp_Y[a] = []
        self.exp_X[a].append(self.get_feature_vector(current_s))
        self.exp_Y[a].append(reward)
```

### Define the environment and training process
- Each action is available to a specific subset of states

```python
def get_not_sampled_states(n_action=10, state_lower_bound=-5):
    not_sampled_states = dict()
    for a in range(0, n_action):
        not_sampled_states[a] = [state_lower_bound+a, state_lower_bound+a+1]
    return not_sampled_states

def shuffle_not_sampled_states(not_sampled_states, n_action):
    random_list = [i for i in range(0, n_action)]
    random.shuffle(random_list)
    
    not_sampled_states_1 = dict()
    for i in range(0, n_action):
        not_sampled_states_1[i] = not_sampled_states[random_list[i]]
    
    return not_sampled_states_1

class env_senario_1():
    def __init__(self, learning_agent, d, n_action=10, n_iteration=100):
        self.n_action = n_action
        self.d = d
        self.states = [i for i in range(-6, 7)]
        self.learning_agent = learning_agent
        self.not_sampled_states = shuffle_not_sampled_states(get_not_sampled_states(n_action), n_action)
        self.n_iter = n_iteration
        
        # if the learning agent is double-q-learning agent, generate a second set of samples for each state
        self.not_sampled_states_1 = shuffle_not_sampled_states(self.not_sampled_states, n_action)
        
    def get_reward(self, s, a):
        return true_action_value_senario_1(s)
        
    def do_train(self):
        agent = self.learning_agent
        not_sampled_states = self.not_sampled_states
        not_sampled_states_1 = self.not_sampled_states_1
        states = self.states
        n_action = self.n_action
        n_iter = self.n_iter
        
        for i_iter in range(0, n_iter):
            for state in states:
                for action in range(0, n_action):
                    reward = self.get_reward(state, action)
                    
                    if agent.get_name() == 'double_q_learning_agent':
                        if state not in not_sampled_states[action]:
                            agent.update_0(state, action, state, reward)
                        if state not in not_sampled_states_1[action]:
                            agent.update_1(state, action, state, reward)
                    else:
                        if state in not_sampled_states[action]:
                            continue
                        agent.update(state, action, state, reward)
        
        agent.compute_weights()
    
    def visualize_result(self):
        agent = self.learning_agent
        
        # plot each actions and max Q(s,a) at each state
        plt.subplot(1, 2, 1)
        plot_true_action_values()
        state_min = -6
        state_max = 6
        x = np.linspace(state_min, state_max, 100)
        max_q = np.array([-9999999999 for _ in x])
        for action in range(0, self.n_action):
            # truncate the q-values in the range of (-4, 4)
            y = np.array([agent.get_q_value(x_i, action) for x_i in x])
            max_q = np.maximum(max_q, y)
            plt.plot(x, y, '-g')
        
        plt.plot(x, max_q, '-b')
        plt.grid()
        plt.gca().set_ylim(-10, 10)
        plt.title('All estimates and max')
        
        # plot the bias
        plt.subplot(1, 2, 2)
        y = np.array([true_action_value_senario_1(x_i) for x_i in x])
        bias = max_q - y
        plt.plot(x, bias, '-r')
        plt.grid()
        plt.title('Bias as function of state')
        plt.gca().set_ylim(-10, 10)
        
        return bias
```

### Test the ordinary q-learning agent

```python
q_learning_agent = q_learning_agent_env_1(d=9, n_action=10)
env = env_senario_1(learning_agent=q_learning_agent, d=9, n_action=10, n_iteration=1)
env.do_train()
bias_q_learning = env.visualize_result()
```

### The Double-Q-Learning agent

```python
class double_q_learning_agent_env_1():
    def __init__(self, d, n_action, learning_rate=0.5, discount_factor=1.0):
        # use n_action functions to estimate the action-state values
        self.weights = None
        self.d = d
        self.n_action = n_action
        self.exp_X_0 = dict()
        self.exp_Y_0 = dict()
        self.exp_X_1 = dict()
        self.exp_Y_1 = dict()
        
    def get_name(self):
        return 'double_q_learning_agent'
        
    def get_feature_vector(self, s):
        return np.array([s**i for i in range(0, self.d)])
    
    def get_q_value(self, s, a):
        w = self.weights[a]
        return np.dot(w, self.get_feature_vector(s))
    
    def compute_weights(self):
        X_0 = self.exp_X_0
        Y_0 = self.exp_Y_0
        weights_0 = np.zeros(shape=(self.n_action, self.d))
        
        X_1 = self.exp_X_1
        Y_1 = self.exp_Y_1
        weights_1 = np.zeros(shape=(self.n_action, self.d))
        
        for a in range(0, self.n_action):
            a_X_0 = np.array(X_0[a])
            a_Y_0 = np.array(Y_0[a])
            weights_0[a, :] = np.dot(np.dot(np.linalg.inv(np.dot(a_X_0.T, a_X_0)), a_X_0.T), a_Y_0)
            
            a_X_1 = np.array(X_1[a])
            a_Y_1 = np.array(Y_1[a])
            weights_1[a, :] = np.dot(np.dot(np.linalg.inv(np.dot(a_X_1.T, a_X_1)), a_X_1.T), a_Y_1)
            
        self.weights = (weights_0 + weights_1)/2
    
    def update_0(self, current_s, a, next_s, reward):
        """
        In this problem, we ignore the next state
        """
        # this is a polynomial function approximation problem, we can't simply use gradient descent
        # so we store the experiences
        if a not in self.exp_X_0:
            self.exp_X_0[a] = []
            self.exp_Y_0[a] = []
        self.exp_X_0[a].append(self.get_feature_vector(current_s))
        self.exp_Y_0[a].append(reward)
             
    def update_1(self, current_s, a, next_s, reward):
        if a not in self.exp_X_1:
            self.exp_X_1[a] = []
            self.exp_Y_1[a] = []
        self.exp_X_1[a].append(self.get_feature_vector(current_s))
        self.exp_Y_1[a].append(reward)
```

```python
double_q_learning_agent = double_q_learning_agent_env_1(d=9, n_action=10)
env = env_senario_1(learning_agent=double_q_learning_agent, d=9, n_action=10, n_iteration=1)
env.do_train()
bias_double_q_learning = env.visualize_result()
```

```python
def plot_compare(bias_double_q, bias_q):
    x = np.linspace(-6, 6, 100)
    
    plt.plot(x, bias_double_q, '-r', label='double-q-learning')
    plt.plot(x, bias_q, '-b', label='q-learning')
    plt.grid()
    plt.legend()
    plt.title('bias comparison')
    plt.gca().set_ylim(-10, 10)

plot_compare(bias_double_q_learning, bias_q_learning)
```

## Senario 2

- This problem is introduced in Sutton & Barto Book and the blog: https://towardsdatascience.com/double-q-learning-the-easy-way-a924c4085ec3
- The problem can be represented as:
<img src="images/overestimation-problem/1.png" alt="mdp" width="800"/>

```python
def true_action_value_senario_2(s):
    values = {'B': -0.5, 'D': 0, 'C': 0, 'A': -0.5}
    return values[s]

def get_available_actions_senario_2(s):
        if s in ['C', 'D']:
            return ['terminate']
        elif s == 'A':
            return ['left', 'right']
        elif s == 'B':
            return ['left0', 'left1', 'left2', 'left3', 'left4', 'left5', 'left6', 'left7', 'left8', 'left9']

def plot_true_action_values_2():
    states = ['A', 'B', 'C', 'D']
    values = dict()
    for s in states:
        values[s] = true_action_value_senario_2(s)
    print('true values: ', values)
    
plot_true_action_values_2()
```

```python
class env_senario_2():
    def __init__(self, learning_agent, n_iteration=100):
        self.states = [i for i in range(-6, 7)]
        self.learning_agent = learning_agent
        self.n_iter = n_iteration
        self.error = None
    
    def get_next_state(self, s, a):
        transit = {('A', 'left'): 'B', ('A', 'right'): 'C', ('C', 'terminate'): 'terminal_state', ('D', 'terminate'): 'terminal_state', ('B', 'left0'): 'D', ('B', 'left1'): 'D', ('B', 'left2'): 'D', ('B', 'left3'): 'D', ('B', 'left4'): 'D', ('B', 'left5'): 'D', ('B', 'left6'): 'D', ('B', 'left7'): 'D', ('B', 'left8'): 'D', ('B', 'left9'): 'D'}
        return transit[(s, a)]
    
    def get_reward(self, s, a):
        if s == 'B':
            return np.random.normal(0, 1.0) - 0.5
        
        rewards = {('A', 'left'): 0, ('A', 'right'): 0, ('C', 'terminate'): 0, ('D', 'terminate'): 0}
        return rewards[(s, a)]
        
    def do_train(self):
        agent = self.learning_agent
        states = self.states
        n_iter = self.n_iter
        
        
        step = 0
        error = []
        available_actions_B = get_available_actions_senario_2('B')
        
        for i_iter in range(0, n_iter):
            # we set B as the initial state because we only experiment on B
            current_state = 'A'
            
            while current_state != 'terminal_state':
                available_actions = get_available_actions_senario_2(current_state)
                action = agent.get_action(current_state, available_actions)
                
                reward = self.get_reward(current_state, action)
                next_state = self.get_next_state(current_state, action)
                
                agent.update(current_state, action, next_state, reward)
                
                error.append(agent.get_value('B', available_actions_B) - true_action_value_senario_2('B'))
                
                current_state = next_state
                step += 1
                
        self.error = error
        return error
    
    def visualize_result(self):
        x = [i for i in range(0, len(self.error))]
        plt.plot(x, self.error, 'r')
```

### The q-learning agent

```python
class q_learning_agent_env_2():
    def __init__(self, learning_rate=0.5, discount_factor=1.0, epsilon=0.1):
        # use n_action functions to estimate the action-state values
        self.q_values = dict()
        self.discount_factor = discount_factor
        self.learning_rate = learning_rate
        self.epsilon = epsilon
        
    def get_name(self):
        return 'q_learning_agent'
        
    def get_action(self, current_state, available_actions, is_random_action=True):      
        if is_random_action and (np.random.uniform(0, 1.0) < self.epsilon):
            return np.random.choice(available_actions)
        
        max_action = None
        max_q_value = None
        for action in available_actions:
            q_value = self.get_q_value(current_state, action)
            if max_q_value is None or q_value > max_q_value:
                max_q_value = q_value
                max_action = action
        return max_action
    
    def get_value(self, s, available_actions):
        if s == 'terminal_state':
            return 0
        
        value = None
        for action in available_actions:
            q_value = self.get_q_value(s, action)
            if value is None or q_value > value:
                value = q_value
        return value
    
    def get_q_value(self, s, a):
        if (s, a) not in self.q_values:
            return 0
        return self.q_values[(s, a)]
    
    def update(self, current_s, a, next_s, reward):
        """
        In this problem, we ignore the next state
        """
        # this is a polynomial function approximation problem, we can't simply use gradient descent
        # so we store the experiences
        if (current_s, a) not in self.q_values:
            self.q_values[(current_s, a)] = reward
            return 
        old_q_value = self.get_q_value(current_s, a)
        new_q_value = old_q_value + self.learning_rate*(reward + self.discount_factor*self.get_value(next_s, get_available_actions_senario_2(next_s)) - old_q_value)
        self.q_values[(current_s, a)] = new_q_value
        
    def print_summary(self):
        print('The learned q-Values: ', self.q_values)
```

### Test the ordinary q-learning agent

```python
def visualize_errors(errors):
    x = [i for i in range(0, len(errors))]
    plt.plot(x, errors, 'r')
    plt.title('bias (overestimation) at state B')
    
def run_exp_q_learning():
    errors = np.zeros(shape=(400, ))
    n_exp = 100

    for _ in range(0, n_exp):
        q_learning_agent_2 = q_learning_agent_env_2(learning_rate=0.05)
        env_2 = env_senario_2(q_learning_agent_2, n_iteration=1000)
        error = env_2.do_train()
        errors += error[:400]

    errors = errors/n_exp
    visualize_errors(errors)
    
    return errors

q_learning_errors = run_exp_q_learning() 
```

### The double q-learning agent

```python
class double_q_learning_agent_0_env_2():
    def __init__(self, learning_rate=0.5, discount_factor=1.0):
        # use n_action functions to estimate the action-state values
        self.q_values_0 = dict()
        self.q_values_1 = dict()
        self.discount_factor = discount_factor
        self.learning_rate = learning_rate
        
    def get_name(self):
        return 'q_learning_agent'
        
    def get_action(self, current_state, available_actions, source=2):        
        max_action = None
        max_q_value = None
        for action in available_actions:
            q_value = self.get_q_value(current_state, action, source=source)
            if max_q_value is None or q_value > max_q_value:
                max_q_value = q_value
                max_action = action
        return max_action
    
    def get_value(self, s, available_actions, source=2):
        if s == 'terminal_state':
            return 0
        
        value = None
        for action in available_actions:
            q_value = self.get_q_value(s, action, source=source)
            if value is None or q_value > value:
                value = q_value
        return value
    
    def get_q_value(self, s, a, source=2):
        if source == 0:
            if (s, a) in self.q_values_0:
                return self.q_values_0[(s, a)]
            else:
                return 0
        if source == 1:
            if (s, a) in self.q_values_1:
                return self.q_values_1[(s, a)]
            else:
                return 0
        
        avg_q = 0
        if (s, a) in self.q_values_0:
            avg_q += self.q_values_0[(s, a)]
        if (s, a) in self.q_values_1:
            avg_q += self.q_values_1[(s, a)]
        return avg_q/2
    
    def update(self, current_s, a, next_s, reward):
        """
        In this problem, we ignore the next state
        """
        # this is a polynomial function approximation problem, we can't simply use gradient descent
        # so we store the experiences
        if np.random.uniform(0,1) < 0.5:
            # update q_value_0
            if (current_s, a) not in self.q_values_0:
                self.q_values_0[(current_s, a)] = reward
                return 
            old_q_value = self.q_values_0[(current_s, a)]
            new_q_value = old_q_value + self.learning_rate*(reward + self.discount_factor*self.get_value(next_s, get_available_actions_senario_2(next_s), source=1) - old_q_value)
            self.q_values_0[(current_s, a)] = new_q_value
        else:
            # update q_value_1
            if (current_s, a) not in self.q_values_1:
                self.q_values_1[(current_s, a)] = reward
                return 
            old_q_value = self.q_values_1[(current_s, a)]
            new_q_value = old_q_value + self.learning_rate*(reward + self.discount_factor*self.get_value(next_s, get_available_actions_senario_2(next_s), source=0) - old_q_value)
            self.q_values_1[(current_s, a)] = new_q_value

    def print_summary(self):
        print('The learned q-Values 0: ', self.q_values_0)
        print('The learned q-Values 1: ', self.q_values_1)
```

```python
def run_exp_double_q_learning():
    errors = np.zeros(shape=(400, ))
    n_exp = 100

    for _ in range(0, n_exp):
        double_q_learning_agent_2 = double_q_learning_agent_0_env_2(learning_rate=0.05)
        env_2 = env_senario_2(double_q_learning_agent_2, n_iteration=1000)
        error = env_2.do_train()
        errors += error[:400]

    errors = errors/n_exp
    visualize_errors(errors)
    
    return errors

double_q_learning_errors = run_exp_double_q_learning() 
```

### Compare two algorithms
- It can be noticed that double q-learning significantly improves the performance

```python
def plot_diff(q_learning_errors, double_q_learning_errors):
    x = [i for i in range(0, len(q_learning_errors))]
    plt.plot(x, q_learning_errors, 'r', label='q-learing bias')
    plt.plot(x, double_q_learning_errors, 'g', label='double q-learing bias')
    plt.legend()
    plt.title('bias (overestimation) at state B')

plot_diff(q_learning_errors, double_q_learning_errors)
```
